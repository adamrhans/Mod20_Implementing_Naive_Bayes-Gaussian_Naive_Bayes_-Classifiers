## Activity Overview

This Jupyter Notebook contains a basic implementation of the gradient descent algorithm. Given the gradient of a function and the choice of an arbitrary starting point, `start`, the gradient descent iteratively moves towards the minimum and returns a point that is at or near the minimum.

Before defining the *function* in Python, let's try and understand the Python *function* below. The `gradient_descent` *function* takes the following for *arguments*: 

- `gradient`: The gradient of the function
- `start`: The starting point of the function from which you want to start the descent towards the local minimum
- `learn_rate`: The learning rate parameter used in the gradient descent algorithm
- `n_iter`: The maximum number of iterations that the gradient descent function will perform
- `tolerance=1e-06`: The error tolerance for the local minimum


```python
import numpy as np

def gradient_descent(
    gradient, start, learn_rate, n_iter=50, tolerance=1e-06
):
    vector = start
    for _ in range(n_iter):
        diff = -learn_rate * gradient(vector)
        #complete the if statement
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
    return vector
```

Now that you have a basic understanding of the implementation of a gradient descent function, it's time for you to test it.

Suppose that you want to find the local minimum of a $f(x) = x^2$ function. This function represents a parabola facing upwards.

Run the code cell below to visualize the $f(x)$ function.

```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
y = x**2

plt.plot(x,y)
plt.plot(0,0, 'ro')
plt.title('Graph of a parabola')
```

In the figure above, the red dot represents the local minimum of the parabola, which in this case, is also called the vertex.

Now that you have defined the function, you can test the implementation of gradient descent by calling the `gradient_descent` *function*.

Note the following: You have set the `gradient` *argument* equal to $2x$ which represents the derivative of $f(x) = x^2$. The starting point, `start`, is set equal to 10 and the learning rate, `learn_rate`, is set equal to 0.2.

```python
gradient_descent(
    gradient=lambda v: 2 * v, start=0.4, learn_rate=0.2
)
```
